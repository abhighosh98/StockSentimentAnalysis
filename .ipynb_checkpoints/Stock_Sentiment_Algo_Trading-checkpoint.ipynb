# Creating a new lexicon from downloaded positive and negative words

Updating Vader Lexicon with External Lexicon

```python
import csv
import pandas as pd

# stock market lexicon
stock_lex = pd.read_csv('lexicon_data/stock_lex.csv')
stock_lex['sentiment'] = (stock_lex['Aff_Score'] + stock_lex['Neg_Score'])/2
stock_lex = dict(zip(stock_lex.Item, stock_lex.sentiment))
stock_lex = {k:v for k,v in stock_lex.items() if len(k.split(' '))==1}
stock_lex_scaled = {}
for k, v in stock_lex.items():
    if v > 0:
        stock_lex_scaled[k] = v / max(stock_lex.values()) * 4
    else:
        stock_lex_scaled[k] = v / min(stock_lex.values()) * -4


positive = []
with open('lexicon_data/lm_positive.csv', 'r') as f:
     reader = csv.reader(f)
     for row in reader:
         positive.append(row[0].strip())
    
negative = []
with open('lexicon_data/lm_negative.csv', 'r') as f:
     reader = csv.reader(f)
     for row in reader:
         entry = row[0].strip().split(" ")
         if len(entry) > 1:
             negative.extend(entry)
         else:
             negative.append(entry[0])

final_lex = {}
final_lex.update({word:2.0 for word in positive})
final_lex.update({word:-2.0 for word in negative})
final_lex.update(stock_lex_scaled)
final_lex.update(sia.lexicon)
sia.lexicon = final_lex
```

# Searches Businesstimes website for given company name. Reads articles and does sentiment analysis

```python
month_dict={
        'Jan' : '01',
        'Feb' : '02',
        'Mar' : '03',
        'Apr' : '04',
        'May' : '05',
        'Jun' : '06',
        'Jul' : '07',
        'Aug' : '08',
        'Sep' : '09', 
        'Oct' : '10',
        'Nov' : '11',
        'Dec' : '12'
}
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time
import pprint

businesstimes_graph_date=[]
businesstimes_graph_sentiment=[]
date_sentiments = {}

for i in range(1,3):
    page = urlopen('https://www.businesstimes.com.sg/search/'+name_company+'?page='+str(i)).read()
    #page = urlopen('https://www.livemint.com/Search/Link/Keyword/infosys').read()
    soup = BeautifulSoup(page, features="html.parser")
    posts = soup.findAll("div", {"class": "media-body"})
    #posts = soup.findAll("div", {"class": "headlineSec"})
    
    for post in posts:
        time.sleep(1)
        url = post.a['href']
        date = post.time.text
        date=str(date[7:]) + "-" + str(month_dict[str(date[3:6])]) + "-" + str(date[0:2])
        print(date, url)
        try:
            link_page = urlopen(url).read()
        except:
            print("Couldn't Read")
            continue
        link_soup = BeautifulSoup(link_page)
        sentences = link_soup.findAll("p")
        passage = ""
        for sentence in sentences:
            passage += sentence.text
        sentiment = sia.polarity_scores(passage)['pos']-sia.polarity_scores(passage)['neg']
        #sentiment = sia.polarity_scores(passage)

        print("%.4f" % sentiment)
        #print(sentiment)
        date_sentiments.setdefault(date, []).append(sentiment)

        date_sentiment = {}

        #for k,v in date_sentiments.items():
        #    date_sentiment[datetime.strptime(k, '%d %b %Y').date() + timedelta(days=1)] = round(sum(v)/float(len(v)),3)

        #earliest_date = min(date_sentiment.keys())
        businesstimes_graph_date.append(date)
        businesstimes_graph_sentiment.append(sentiment)
        
        #earliest_date = datetime.datetime(2020,1,1)

        #print(date_sentiment)
```

# Searches Livemint website for given company name. Reads articles and does sentiment analysis

```python
livemint_graph_date=[]
livemint_graph_sentiment=[]
page = urlopen('https://www.livemint.com/Search/Link/Keyword/'+name_company).read()
soup = BeautifulSoup(page, features="html.parser")
posts = soup.findAll("div", {"class": "headlineSec"})
for post in posts:
        time.sleep(1)
        date_time=re.search(r'(?<=data-expandedtime=)[^,]*',str(post))
        date_time=date_time.group(0)[10:]
        url = post.a['href']
        url='https://www.livemint.com'+url
        date=str(date_time[7:]) + "-" + str(month_dict[str(date_time[3:6])]) + "-" + str(date_time[0:2])
        print(date,url)
        
        try:
            link_page = urlopen(url).read()
        except:
            print("Couldn't read")
        link_soup = BeautifulSoup(link_page)
        sentences = link_soup.findAll("p")
        passage = ""
        for sentence in sentences:
            passage += sentence.text
        sentiment = sia.polarity_scores(passage)['pos']-sia.polarity_scores(passage)['neg']
        #sentiment = sia.polarity_scores(passage)
        print("%.4f" % sentiment)
        #print(sentiment)
        livemint_graph_date.append(date)
        livemint_graph_sentiment.append(sentiment)
        
                
```

```python
from yahoofinancials import YahooFinancials
```

```python
yahoo_financials = YahooFinancials(ticker_symbol)

balance_sheet_data_qt = yahoo_financials.get_financial_stmts('annual', 'balance')
```

```python
balance_sheet=balance_sheet_data_qt['balanceSheetHistory']
balance_sheet=balance_sheet[ticker_symbol]

```

```python
list_of_headings=[]
list_of_years=[]
temp_list=[]
list_of_crores=[]
for y in balance_sheet:
  #print(y)
  for k in y.keys():
    list_of_years.append(k)
    for heading in y[k]:
      list_of_headings.append(heading)
  
  for num in y.values():
    #print(num)
    temp_list=[]
    for crores in num.values():
      temp_list.append(crores)
    #print(temp_list)

    list_of_crores.append(temp_list)
  


```

```python
temp_dict={
    'Headings/Years': list(set(list_of_headings)),
    list_of_years[0]: list_of_crores[0],
    list_of_years[1]: list_of_crores[1],
    list_of_years[2]: list_of_crores[2],
    list_of_years[3]: list_of_crores[3],
}
print(temp_dict)
```
